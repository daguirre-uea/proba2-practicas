## Preparación del entorno


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats, integrate

# Ajustes de visualización
plt.rcParams['figure.figsize'] = (8,5)
plt.rcParams['font.size'] = 12
```

# 1. Distribución Uniforme Continua

Una variable aleatoria continua \(X\) tiene distribución uniforme en el intervalo \([a,b]\) si su función de densidad es:

$$
f_X(x)=
\begin{cases}
\frac{1}{\beta-\alpha}, & \alpha \le x \le \beta \\
0, & \text{en otro caso}
\end{cases}
$$

La distribución uniforme asigna la misma probabilidad a todos los subintervalos de igual longitud.

## 1.1 PDF de una distribución uniforme

```python
alfa = 2
beta = 8

X = stats.uniform(loc=alfa, scale=beta-alfa)

x = np.linspace(0,10,1000)
pdf = X.pdf(x)

plt.plot(x, pdf, lw=2)
plt.fill_between(x, pdf, alpha=0.3)

plt.title('PDF - Distribución Uniforme')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

## 1.2 CDF de una distribución uniforme

La función acumulada está dada por:

$$
F_X(x)=P(X\le x)
$$

y representa la probabilidad acumulada hasta el valor $x$.


```python
cdf = X.cdf(x)

plt.plot(x, cdf, lw=2)

plt.title('CDF - Distribución Uniforme')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid()
plt.show()
```

## 1.3 Cálculo de probabilidades

Calcular:

$$
P(3 \le X \le 6)
$$


```python
p = X.cdf(6) - X.cdf(3)

print("P(3 <= X <= 6) =", p)
```

## 1.4 Esperanza y varianza

Para una distribución uniforme:

$$
\mathbb{E}[X]=\frac{a+b}{2}
$$

$$
Var(X)=\frac{(b-a)^2}{12}
$$


```python
esperanza = X.mean()
varianza = X.var()

print("Esperanza =", esperanza)
print("Varianza =", varianza)
```

## 1.5 Problema de aplicación

Un autobús pasa cada 10 minutos de manera uniforme. Una persona llega aleatoriamente a la parada.

Sea $X$ el tiempo de espera en minutos.

1. ¿Cuál es la probabilidad de esperar menos de 4 minutos?
2. ¿Cuál es el tiempo promedio de espera?

```python
X_bus = stats.uniform(loc=0, scale=10)

p_espera = X_bus.cdf(4)
promedio = X_bus.mean()

print("Probabilidad de esperar menos de 4 minutos:", p_espera)
print("Tiempo promedio de espera:", promedio)
```

# 2. Distribución Exponencial

La distribución exponencial modela tiempos de espera entre eventos.

Su función de densidad es:

$$
f_X(x)=\lambda e^{-\lambda x}, \quad x \ge 0
$$

donde $\lambda$ es la tasa del proceso.


## 2.1 PDF de una distribución exponencial

```python
lam = 0.5

X_exp = stats.expon(scale=1/lam)

x = np.linspace(0,15,1000)
pdf_exp = X_exp.pdf(x)

plt.plot(x, pdf_exp, lw=2)
plt.fill_between(x, pdf_exp, alpha=0.3)

plt.title('PDF - Distribución Exponencial')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

## 2.2 CDF de una distribución exponencial

```python
cdf_exp = X_exp.cdf(x)

plt.plot(x, cdf_exp, lw=2)

plt.title('CDF - Distribución Exponencial')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid()
plt.show()
```

## 2.3 Cálculo de probabilidades

Calcular:

$
P(X > 3)
$


```python
p = 1 - X_exp.cdf(3)

print("P(X > 3) =", p)
```

## 2.4 Esperanza y varianza

Para una distribución exponencial:

$$
\mathbb{E}[X]=\frac{1}{\lambda}
$$

$$
Var(X)=\frac{1}{\lambda^2}
$$

```python
esperanza_exp = X_exp.mean()
varianza_exp = X_exp.var()

print("Esperanza =", esperanza_exp)
print("Varianza =", varianza_exp)
```

## 2.5 Problema de aplicación

El tiempo de vida de un sensor electrónico sigue una distribución exponencial con tasa:

$$
\lambda = 0.2
$$

1. ¿Cuál es la probabilidad de que el sensor dure más de 8 horas?
2. ¿Cuál es la vida útil promedio?


```python
lam_sensor = 0.2

X_sensor = stats.expon(scale=1/lam_sensor)

p_dure = 1 - X_sensor.cdf(8)
vida_promedio = X_sensor.mean()

print("Probabilidad de durar más de 8 horas:", p_dure)
print("Vida promedio:", vida_promedio)
```
